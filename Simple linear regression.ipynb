Simple linear regression

```python
import numpy as np

import matplotlib.pyplot as plt

import sklearn
```

```python
# data

x = np.array([1,2,3,4])

print(x)

y = np.array([1,3,5,7])

print(y)
```

```python
x.shape
```

```python
x = x.reshape(-1, 1)
```

```python
plt.plot(x, y, 'o')

plt.show()
```

```python
# model 

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x, y)
```

```python
# y = m * x + c 

m = model.coef_[0]

c = model.intercept_

print(f'm = {m}, c = {c}')
```

```python
plt.plot(x, y, 'o')

plt.plot(x, model.predict(x), '-')
```

```python
result = model.predict([[5]])
                       
result
```
